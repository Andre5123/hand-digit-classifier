Use this to plot loss curves and confusion matrix after training.


```python
import os
import sys
sys.path.append("../src")

import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from pathlib import Path

from data.preprocess import make_classification_dataset
```

Check if classifier model is working on a few examples

```python
model = tf.keras.models.load_model("../models/classifier_best.keras")

test_dataset, test_size = make_classification_dataset("../data/raw/digits/test", 32)

for images, labels in test_dataset.take(1):
    predictions = model.predict(images)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = labels.numpy().astype(int)
    print("predicted: ", predicted_classes)
    print("true: ", true_classes)
    confusion_matrix = tf.math.confusion_matrix(true_classes, predicted_classes)
print(confusion_matrix) 
```
